## 1. Code Refactoring
### 1.1 Ubah gaya penulisan kode program berikut sesuai dengan standar PEP8

```python
#Program untuk menghitung luas dan keliling lingkaran
PI = 3.14  #1 Seharusnya Phi dan upper case

def hitung_luas_lingkaran(jari_jari):  
    luas = PI * jari_jari**2  #2 Nama konstanta harus uppercase
    return luas #24 Nama variable belum jelas

def hitung_keliling_lingkaran(jari_jari): #3 Nama fungsi harus snakecase
    keliling = 2 * PI * jari_jari  #4 Nama konstanta harus uppercase
    return keliling #23 Nama variale belum jelas

def tampilkan_menu():
    print('')
    print("----------------------------------------------------")
    print('Program untuk menghitung luas dan keliling lingkaran') #5 Kelebihan whitespace
    print('Mau ngitung apa nih bre?') #6 Kelebihan whitespace
    print('1. Luas Lingkaran') #7 Kelebihan whitespace
    print('2. Keliling Lingkaran') #8 Kelebihan whitespace
    print('3. Stop Program') #9 Kelebihan whitespace
    print('')


pilihan = 'bukan 3' #10 Logika aneh
while pilihan != 3: #22 Nama variable belum jelas
    tampilkan_menu()
    pilihan = input ('Pilihan : ') #11 Kelebihan spasi sesudah input dan nama variable belum jelas

    if pilihan== '1': #12 Kurang whitespace dan nama variable belum jelas
        jari_jari = input('Jari Jari : ')
        jari_jari = float(jari_jari)
        hasil = hitung_luas_lingkaran(jari_jari) #13 Nama fungsi harus snakecase
        print("Luas Lingkaran: ", hasil )

    elif pilihan == '2': #21. Nama variable harus jelas dan kurang spasi setelah operator
        jari_jari= input('Jari Jari:') #14 Kurang whitespace di depan input dan di depan variable
        jari_jari= float(jari_jari) #15 Kurang whitespace di depan float dan dibelakang variable
        hasil = hitung_keliling_lingkaran(jari_jari) #16 Kurang whitespace, salah penamaan variable dan nama fungsi harus snakecase
        print("Keliling Lingkaran : ", hasil) #17 Kelebihan whitespace dan salah penamaan variable

    elif pilih == '3': #18 Masing - masing diberi spasi sebelum dan sesudah operator
        print("TERIMA KASIH SUDAH MENGGUNAKAN PROGRAM KAMI") #19. Kelebihan whitespace
        break

    else:
        print('Pilihan tidak tersedia')
        tampilkan_menu() #20. Kelebihan whitespace
```

#Enumerate [Sumber](https://www.geeksforgeeks.org/enumerate-in-python/)
Seringkali, ketika berhadapan dengan iterator, kita juga perlu menghitung iterasi. Python memudahkan tugas programmer dengan menyediakan fungsi built-in enumerate() untuk tugas ini.
Metode Enumerate() menambahkan penghitung ke iterable dan mengembalikannya dalam bentuk objek enumerasi. Objek enumerated ini kemudian dapat digunakan secara langsung untuk loop atau diubah menjadi daftar tupel menggunakan metode list().

```python
# Kunci Jawaban
pemain_dan_medali_sorted.extend([[2,"Rio"], [5, "Dewi"]])
pemain_dan_medali_sorted.sort()
pemain_dan_medali_sorted.reverse()

def seleksi_tim_sudirman(list_pemain_dan_medali):
  for index, (medali, pemain) in enumerate(list_pemain_dan_medali):
    if index <= 13:
      print(f'{pemain} layak bergabung di Tim Sudirman')
    else:
      print(f'{pemain} belum layak bergabung di Tim Sudirman')
        
seleksi_tim_sudirman(pemain_dan_medali_sorted)
```
